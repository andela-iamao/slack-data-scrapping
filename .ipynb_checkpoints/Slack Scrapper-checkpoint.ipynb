## Most frequent words in channels history

```python
general_history = sc.api_call('channels.history', channel='C0R29KRLK', count=200)
```

```python
len(general_history['messages'])
```

```python
all_messages = []

for message in general_history['messages']:
    if 'text' in message.keys():
        all_messages.append(message['text'])
```

```python
# print(all_messages)
for message in all_messages:
    print(message + '\n')
```

## Text preprocessing

Remove non alphanumeric characters

```python
import re

# Strip whitespace first

for index in range(len(all_messages)):
    all_messages[index].strip()
    all_messages[index] = re.sub('[^0-9a-zA-Z\s]+', '', all_messages[index])
```
